```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
import math
from PIL import Image as im
```

```python
#No 2C
def convolution2d(image, kernel, stride, padding):

  # add paddinng
  image = np.pad(image, padding, mode='constant', constant_values=0)

  # pick img input width and height values including insertion pad
  kernel_height, kernel_width = kernel.shape
  padded_height, padded_width = image.shape

  """
  Formula to find the output size using convolution:
    [W-K+2SP/S]+1
  W = input img size (width/height)
  K = kernel size
  P = padding size
  S = stride value


  """
  output_height = (padded_height - kernel_height) // stride + 1
  output_width = (padded_width - kernel_width) // stride + 1

  new_image = np.zeros((output_height, output_width)).astype(np.float32)

  # convolution process
  for y in range(0, output_height):
    for x in range(0, output_width):
      new_image[y][x] = np.sum(image[y * stride:y * stride + kernel_height, x * stride:x * stride + kernel_width] * kernel).astype(np.float32)
  return new_image
```

```python
#No 2D
img = cv.imread('/content/drive/MyDrive/PCVK/images/mandrill.tiff')
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#No 2E
# sharpen
kernal_sharpen = np.array([[0,-1,0],
                           [-1,5,-1],
                           [0,-1,0]])
# emboss
kernal_emboss = np.array([[-2,-1,0],
                          [-1,1,1],
                          [0,1,2]])


# left sobel edge detection
kernal_sobel = np.array([[1,0,-1],
                         [2,0,-2],
                         [1,0,-1]])


# canny edge detection
kernal_canny = np.array([[-1,-1,-1],
                         [-1,8,-1],
                         [-1,-1,-1]])


# Average
kernal_average = np.array([[0.11,0.11,0.11],
                         [0.11,0.11,0.11],
                         [0.11,0.11,0.11]])

# Low Pass
kernal_lowpass = np.array([[0.0833,0.0833,0.0833],
                         [0.0833,0.333,0.0833],
                         [0.0833,0.0833,0.0833]])

# High Pass
kernal_highpass = np.array([[-1,0,1],
                         [-1,0,3],
                         [-3,0,1]])
```

```python
cv2_imshow(convolution2d(img_gray,kernal_sharpen,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_emboss,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_sobel,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_canny,1,2))
```

```python
# 21x21 Gaussian Blur
kernel_size = 21
sigma = math.sqrt(kernel_size)
gaussian_kernel = cv.getGaussianKernel(kernel_size, sigma)
gauss_kernel = gaussian_kernel @ gaussian_kernel.transpose()

cv2_imshow(convolution2d(img_gray,gauss_kernel,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_average,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_lowpass,1,2))
```

```python
cv2_imshow(convolution2d(img_gray,kernal_highpass,1,2))
```
