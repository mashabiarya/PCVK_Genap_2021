```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
ImagePath = '/content/drive/MyDrive/PCVK/asukauas.png';

originalmage = cv.imread(ImagePath)
originalmage = cv.cvtColor(originalmage, cv.COLOR_BGR2RGB)

ReSized1 = cv.resize(originalmage, (1280, 820))

plt.imshow(ReSized1, cmap='gray')
```

```python
GSImg = cv.cvtColor(originalmage, cv.COLOR_BGR2GRAY)
ReSized2 = cv.resize(GSImg, (1280, 820))
plt.imshow(ReSized2, cmap='gray')
```

```python
smooth = cv.medianBlur(GSImg, 5)
ReSized3 = cv.resize(smooth, (1280, 820))
plt.imshow(ReSized3, cmap='gray')
```

```python
getEdge = cv.adaptiveThreshold(smooth, 255, 
  cv.ADAPTIVE_THRESH_MEAN_C, 
  cv.THRESH_BINARY, 9, 9)
ReSized4 = cv.resize(getEdge, (1280, 820))
plt.imshow(ReSized4, cmap='gray')
```

```python
colorImage = cv.bilateralFilter(originalmage, 9, 300, 300)
ReSized5 = cv.resize(colorImage, (1280, 820))
plt.imshow(ReSized5, cmap='gray')
```
